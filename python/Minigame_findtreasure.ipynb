```python
import random
import matplotlib.pyplot as plt

treasure_position=[0,0]
player_position=[0,0]
count=0
#격자 그리기
def show_grid(n):
    plt.xlim(0, n)
    plt.ylim(0, n)

    # 눈금 간격 1로 변경(x, y축 모두)
    plt.xticks(range(n+1))
    plt.yticks(range(n+1))

    plt.grid()

    plt.show()

# 게임 초기화
def initialize_game(board_size):
    player_position[0]=random.randrange(board_size+1)
    player_position[1]=random.randrange(board_size+1)
    treasure_position[0]=random.randrange(board_size+1)
    treasure_position[1]=random.randrange(board_size+1)

# 거리 계산
def calculate_distance(treasure_position, player_position):
    return abs(player_position[0]-treasure_position[0])+abs(player_position[1]-treasure_position[1])


# 플레이어 이동
def move_player(board_size, player_position, direction):
    global count
    if direction=="동" and player_position[0]<board_size:
        player_position[0]+=1
        count+=1
    elif direction=="서" and player_position[0]>0:
        player_position[0]-=1
        count+=1
    elif direction=="북" and player_position[1]<board_size:
        player_position[1]+=1
        count+=1
    elif direction=="남" and player_position[1]>0:
        player_position[1]-=1
        count+=1
    else:
        print("가능 범위 외 움직임입니다. 다시 이동하세요")

# 게임 실행
def play_game(board_size):
    show_grid(board_size)
    initialize_game(board_size)
    while True:
        a=input("이동 방향을 입력하세요(동,서,남,북,종료)")
        if a=="동" or a=="서" or a=="남" or a=="북":
            move_player(board_size,player_position,a)
            print(f"{a}쪽으로 한칸 이동! 남은 거리 : {calculate_distance(treasure_position,player_position)}")
            if calculate_distance(treasure_position,player_position)==0:
                print(f"보물찾기 성공! 이동 횟수 : {count}")
                break
        elif a=="종료":
            break
    

# 게임 보드 크기 설정 및 게임 시작
if __name__ == "__main__":
    board_size = 5  # 보드 크기를 5x5로 설정
    
    
    play_game(board_size)
```
